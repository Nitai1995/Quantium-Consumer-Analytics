```python
store_df.info()
```

```python
store_df['DATE'] = pd.to_datetime(store_df['DATE'])
store_df['MONTH'] = pd.to_datetime(store_df['DATE'].dt.strftime('%m-%Y'))
store_df
```

```python
df13 = store_df.groupby('MONTH',as_index=False)['TOT_SALES'].sum()
plt.subplots(figsize=(12,8))
plt.plot(df13['MONTH'], df13['TOT_SALES'])
```

#### 2. Total Sales Revenue for each trial store:

```python
df6 = store_df.groupby('STORE_NBR', as_index=False)['TOT_SALES'].sum()
df6.loc[df6['STORE_NBR'].isin([77,86,88])]
```

#### 3. Total number of customers for each trial store:

```python
df7 = store_df[store_df['STORE_NBR'].isin([77,86,88])]#['LYLTY_CARD_NBR'].value_counts()
#df3 = store_df.groupby('STORE_NBR', as_index=False)
#df3[df3['STORE_NBR'].isin([77,86,88, 266])]
df8 = pd.DataFrame({'STORE_NBR':[77,86,88],
                   'NMBR_OF_CSTMRS':[
                       len(df7[df7['STORE_NBR']==77]['LYLTY_CARD_NBR'].unique()),
                       len(df7[df7['STORE_NBR']==86]['LYLTY_CARD_NBR'].unique()),
                       len(df7[df7['STORE_NBR']==88]['LYLTY_CARD_NBR'].unique())
                   ]})
df8
```

#### 4. Total number of transactions in each trial store:

```python
df9 = store_df.groupby('STORE_NBR', as_index=False)['TXN_ID'].count()
#df4
df9[df9['STORE_NBR'].isin([77,86,88])]
```

#### 5. Average number of transactions per customer in each trial store:

```python
df10 = store_df.groupby(['STORE_NBR','LYLTY_CARD_NBR'],as_index=False)['TXN_ID'].count()
#df5
df11 = pd.DataFrame({'STORE_NBR':[77,86,88],
                   'AVG_TRNSCTN':[
                       df10[df10['STORE_NBR'] == 77]['TXN_ID'].mean(),
                       df10[df10['STORE_NBR'] == 86]['TXN_ID'].mean(),
                       df10[df10['STORE_NBR'] == 88]['TXN_ID'].mean()
                   ]})
df11
```

#### 6. Brand wise Sales in each trial store:

```python
df12 = store_df.groupby(['STORE_NBR','BRAND'], as_index=False)['PROD_QTY','TOT_SALES'].sum()
df12[df12['STORE_NBR'].isin([77,88,86])]
```

#### 7. Total Number of Brands in each trial store:

```python
df13 = pd.DataFrame({'STORE_NBR':[77,86,88],
                   'BRAND':[
                       len(df7[df7['STORE_NBR']==77]['BRAND'].unique()),
                       len(df7[df7['STORE_NBR']==86]['BRAND'].unique()),
                       len(df7[df7['STORE_NBR']==88]['BRAND'].unique())
                   ]})
df13
```
